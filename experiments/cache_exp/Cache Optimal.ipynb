```python
import numpy as np

# Settings

Vdd = 3
I_tx = 11.6e-3
I_sleep = 80e-9
I_idle = 396e-6
I_rx = 12.7e-3
I_sleep_idle = 691e-12
I_idle_tx = 6.63e-6
I_idle_rx = 6.63e-6
P_tx = I_tx * Vdd
P_sleep = I_sleep * Vdd
P_idle = I_idle * Vdd
P_rx = I_rx * Vdd

Energy = Vdd * 3 * 3600

# Communications settings

R = 250000
L_get = 87 * 8
L_Ans = 96 * 8
S_get = L_get / R
S_Ans = L_Ans / R
S_ack = 0.608e-3
S_get_Rxtx = S_ack
S_Ans_Rxtx = S_ack
T_period = 0.244
T_d = 0.16e-3
T_p = 0.4e-3

# Simulation settings

N = 12
l = 0.5
cmin, cmax = 1, 9


t_act = 4e-4
t_sleep = T_period - t_act

S_get_Rxrx = (3 * S_get) / 2 + T_p
S_Ans_Rxrx = (3 * S_Ans) / 2 + T_p
S_Ans_Txrx = (3 + np.floor((t_sleep - S_Ans) / S_Ans)) * T_d / 2 + S_ack
S_get_Txrx = (3 + np.floor((t_sleep - S_get) / S_get)) * T_d / 2 + S_ack
S_Ans_Txtx = (3 + np.floor((t_sleep - S_Ans) / S_Ans)) * S_Ans / 2
S_get_Txtx = (3 + np.floor((t_sleep - S_get) / S_get)) * S_get / 2


gamma_rx_get = P_sleep * (3 * S_get / 2 + T_p + S_ack) / T_period
gamma_rx_Ans = P_sleep * (3 * S_Ans / 2 + T_p + S_ack) / T_period


gamma_tx_get = P_sleep * ((3 + np.floor((t_sleep - S_get) / S_get)) / 2
* (S_get + T_d) + S_ack) / T_period

gamma_tx_Ans = P_sleep * ((3 + np.floor((t_sleep - S_Ans) / S_Ans)) / 2
* (S_Ans + T_d) + S_ack) / T_period

omega_sleep = P_sleep * t_sleep / T_period
- gamma_tx_Ans - gamma_rx_get - gamma_tx_get - gamma_rx_Ans

omega_cl = t_act * P_rx / T_period

def ratio(c, t):
    if t <= c:
        r = np.ceil(c / t) * t
    else:
        r = t
    return r

t = 1 / l

for c in np.arange(1.05, 9.5, 1.05):
    r_i = ratio(c, t)

    omega_CoAP_tx = (P_tx * S_Ans_Txtx + P_rx * S_Ans_Txrx) / r_i
    + (P_tx * S_Ans_Txtx + P_rx * S_Ans_Txrx) / r_i
    + (P_tx * S_Ans_Txtx + P_rx * S_Ans_Txrx) / r_i

    omega_CoAP_rx = (P_tx * S_get_Rxrx + P_rx * S_get_Rxtx) / r_i
    + (P_tx * S_get_Rxrx + P_rx * S_get_Rxtx) / r_i
    + (P_tx * S_get_Rxrx + P_rx * S_get_Rxtx) / r_i

    omega_root_tx = (P_tx * S_get_Txtx + P_rx * S_get_Txrx) / r_i
    + (P_tx * S_get_Txtx + P_rx * S_get_Txrx) / r_i

    omega_root_rx = (P_tx * S_Ans_Rxrx + P_rx * S_Ans_Rxtx) / r_i
    + (P_tx * S_Ans_Rxrx + P_rx * S_Ans_Rxtx) / r_i

    omega_tot = omega_sleep + omega_cl + omega_root_tx + omega_root_tx
    + omega_CoAP_tx + omega_CoAP_rx

    life_min = Energy / omega_tot
    c
    life_min_days = Energy / omega_tot / (3600 * 24)
    print(life_min_days)
```

```python
from deap.base import Toolbox
from deap.tools import selNSGA2, mutPolynomialBounded, cxSimulatedBinaryBounded, initRepeat
from deap.algorithms import eaSimple


NGEN = 100
BOUND_LOW = 1
BOUND_UP = 4
NDIM = 12
POPULATION_SIZE = 100
IND_SIZE=10

toolbox = Toolbox()


creator.create("FitnessMax", base.Fitness, weights=(1.0,))
creator.create("Individual", list, fitness=creator.FitnessMax)



toolbox.register("attr_float", np.random.uniform, 1, 9)
toolbox.register("individual", tools.initRepeat, creator.Individual,
                 toolbox.attr_float, n=IND_SIZE)
toolbox.register("population", tools.initRepeat, list, toolbox.individual)

# On utilise la même méthode de sélection que dans le papier
toolbox.register("select", selNSGA2)

# Executes a simulated binary crossover that modify in-place the input individuals.
# The simulated binary crossover expects sequence individuals of floating point numbers.
toolbox.register("mate", cxSimulatedBinaryBounded, low=BOUND_LOW, up=BOUND_UP, eta=20.0)

# Mutation classique en NSGA2
toolbox.register("mutate", mutPolynomialBounded, low=BOUND_LOW, up=BOUND_UP, eta=20.0, indpb=1.0/NDIM)

pop = toolbox.population(n=100)

result = eaSimple(pop, toolbox, cxpb=0.5, mutpb=0.2, ngen=10, verbose=False)
```
